## 6. Introduction to Simple Linear Regression

In applied economics, we often study the relationship between two variables.

For example:
- Wage (y)
- Education (x)

The simple linear regression model is:

$$
y_i = \beta_0 + \beta_1 x_i + e_i
$$

Where:

- $\beta_0$ = intercept  
- $\beta_1$ = slope coefficient  
- $e_i$ = error term  

The slope estimator formula is:

$$
\hat{\beta}_1 =
\frac{\sum (x_i - \bar{x})(y_i - \bar{y})}
{\sum (x_i - \bar{x})^2}
$$


### 6.1 Simple Numerical Example

```python
import numpy as np

# Example data
x = np.array([1, 2, 3, 4], dtype=float)
y = np.array([2, 4, 5, 8], dtype=float)

x_bar = x.mean()
y_bar = y.mean()

beta1_hat = ((x - x_bar)*(y - y_bar)).sum() / ((x - x_bar)**2).sum()
beta0_hat = y_bar - beta1_hat*x_bar

print("Estimated intercept (beta0):", beta0_hat)
print("Estimated slope (beta1):", beta1_hat)
```

## 7. Estimating the Model Using statsmodels

We can also estimate the same model using the statsmodels library.

Official documentation:  
https://www.statsmodels.org/stable/index.html


```python
import statsmodels.api as sm

X = sm.add_constant(x)
model = sm.OLS(y, X).fit()

print(model.summary())
```
