### Interactive 3D Surface Plot for a 2D NumPy Matrix
This section visualizes a two-dimensional NumPy matrix using an interactive 3D surface plot.

Each matrix element is represented as a height value (z-axis), while row and column indices define the x and y axes. This approach helps interpret matrix data in a spatial and intuitive way.

For a detailed explanation of [Plotly Surface Plots](https://www.youtube.com/watch?v=9GYmFXBitBw&list=PLBSCvBlTOLa8rf2kGkP_Bx5xXqT-er4Yq/)

```python
import numpy as np
import plotly.graph_objects as go   # Plotly for interactive visualization

# Create a 2D NumPy matrix
two_dimension_vector = np.array([[1, 2, 3, 4],
                                 [5, 6, 7, 8]])

# Create coordinate grids for X and Y axes
x = np.arange(two_dimension_vector.shape[1])  # Column indices
y = np.arange(two_dimension_vector.shape[0])  # Row indices
X, Y = np.meshgrid(x, y)  # Generate coordinate matrix

# Create interactive 3D surface plot
fig = go.Figure(data=[go.Surface(
    z=two_dimension_vector,  # Z values (matrix values)
    x=X,                     # X coordinates
    y=Y                      # Y coordinates
)])

# Update layout for better visualization
fig.update_layout(
    title="Interactive 3D Surface Plot of a 2D NumPy Matrix",
    scene=dict(
        xaxis_title="Columns",
        yaxis_title="Rows",
        zaxis_title="Matrix Values"
    )
)

# Display interactive plot
fig.show()
```

## Applying a Quadratic Function on a NumPy Matrix and Finding the Maximum Value
### Loop-Based vs Vectorized Computation in NumPy (Quadratic Function Example)
In this section, we compare two approaches for applying a quadratic function

$$
f(x) = 5x^2 + 3
$$

to a NumPy matrix:

1. A loop-based implementation (explicit iteration).
2. A vectorized implementation (NumPy optimized computation).

The goal is to demonstrate the computational efficiency difference between pure Python loops and NumPy’s vectorized operations.

### Further Reading

For an intuitive explanation of replacing explicit loops with vectorized operations in Python:

[Ditch loops and Embrace vectorization (Medium)](https://medium.com/@vinodkumargr/ditch-loops-and-embrace-vectorization-25e578a5a4b7)

### 1) Loop-Based Approach (Explicit Iteration)
This implementation applies the function element-by-element using nested loops and manually tracks the maximum value.

```python
import numpy as np  # NumPy library is used for numerical operations and matrix handling

# Example matrix (2 rows, 3 columns)
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])  # Creates a 2D NumPy array

# Function definition: y = 5x^2 + 3
def f(x):
    return 5*x**2 + 3  # Quadratic function

max_y = -np.inf  # Initialize maximum y as negative infinity (ensures first comparison works)
max_x = None     # Variable to store the x value that produces the maximum y

# Iterate through each row in the matrix
for row in matrix:
    # Iterate through each element in the current row
    for x in row:
        y = f(x)  # Apply the function to current element
        print(f"x = {x}, y = {y}")  # Print each calculation

        if y > max_y:  # Check if current y is greater than stored maximum
            max_y = y  # Update maximum y
            max_x = x  # Store corresponding x value

print("\nMaximum value:")
print("x =", max_x)  # x that gives the highest y
print("y =", max_y)  # Highest y value
```

### 2) Vectorized Approach (NumPy Array Operations)
This implementation applies the function to the entire matrix at once using NumPy vectorized operations, then finds the maximum efficiently.

```python
import numpy as np

# Example matrix (2 rows, 3 columns)
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

# Vectorized function application: f(x) = 5x^2 + 3 for all elements
y = 5 * matrix**2 + 3

print("y matrix (function values):\n", y)

# Maximum y value
max_y = np.max(y)

# Get 2D index of maximum
max_index = np.unravel_index(np.argmax(y), y.shape)    # np.unravel_index takes a flat index (from np.argmax) and returns its corresponding 2D position (row, column).

# Get corresponding x value
max_x = matrix[max_index]

print("\nMaximum value:")
print("x =", max_x)
print("y =", max_y)
```

### 3) Benchmark: Loop vs Vectorized Performance
We compare execution time for the loop-based and vectorized approaches using a larger random matrix.

```python
import numpy as np
import time

# Create a large matrix for performance comparison
#matrix = np.array([[1, 2, 3],
#                   [4, 5, 6]])

matrix = np.random.rand(500, 500)
# ==================================================
# 1) Loop-Based Implementation
# ==================================================
start_time = time.time()

max_y_loop = -np.inf
max_x_loop = None

for row in matrix:
    for x in row:
        y_val = 5*x**2 + 3
        if y_val > max_y_loop:
            max_y_loop = y_val
            max_x_loop = x

loop_time = time.time() - start_time

# ==================================================
# 2) Vectorized Implementation
# ==================================================
start_time = time.time()

# Apply function to entire matrix at once
y_vec = 5 * matrix**2 + 3

# Maximum y value
max_y_vec = np.max(y_vec)

# Convert flat index (from argmax) into 2D index
max_index_vec = np.unravel_index(np.argmax(y_vec), y_vec.shape)

# Get corresponding x value
max_x_vec = matrix[max_index_vec]

vectorized_time = time.time() - start_time

# ==================================================
# Results
# ==================================================
print("Loop Version Time (seconds):", loop_time)
print("Vectorized Version Time (seconds):", vectorized_time)

print("\nLoop Max:")
print("x =", max_x_loop)
print("y =", max_y_loop)

print("\nVectorized Max:")
print("x =", max_x_vec)
print("y =", max_y_vec)
```

## Solving a System of Linear Equations Using NumPy

Consider the following system of linear equations:

$$
x + 2y = 1
$$

$$
4x + 5y = 1
$$

This system can be written in matrix form as:

$$
A \cdot V = d
$$

Where:
$$
A =
\begin{bmatrix}
1 & 2 \\
4 & 5
\end{bmatrix}
$$

$$
V =
\begin{bmatrix}
x \\
y
\end{bmatrix}
$$

$$
d =
\begin{bmatrix}
1 \\
1
\end{bmatrix}
$$


---

### Why do we separate the system as A, V, and d?

- **A** is the coefficient matrix.  
  It contains the numerical coefficients of the variables.

- **V** is the variable (unknown) vector.  
  It represents the values we want to solve for (x and y).

- **d** is the constant vector.  
  It contains the values on the right-hand side of the equations.

This matrix representation allows us to solve the system using linear algebra methods instead of solving the equations manually.

---

### Mathematical Solution Concept

If matrix (A) is invertible, the solution can be written as:

$$
V = A^{-1} d
$$

However, instead of explicitly computing the inverse (which is computationally expensive and numerically unstable), NumPy provides:

$$
\texttt{np.linalg.solve(A, d)}
$$
or
$$
\texttt{np.dot(np.linalg.inv(A), d)}
$$

These functions directly computes the solution using efficient linear algebra algorithms ([LU decomposition](https://en.wikipedia.org/wiki/LU_decomposition)).

---
This approach scales efficiently to larger systems and is fundamental in scientific computing, econometrics, optimization, and machine learning.

```python
A = np.array([[1,2],[4,5]])
d = np.array([1,1])

#V = np.linalg.solve(A,d)    # Solves the linear system A @ V = d and returns the solution vector V.
V = np.linalg.inv(A) @ d     # If you want to use the inverse, the correct multiplication operator is @ or np.dot((np.linalg.inv(A),d)
V
#note: In Python, the @ operator performs matrix multiplication.


```

```python
np.linalg.solve(A,d) # It takes matrix A and result vector d, and calculates the values of the variables automatically.
```

### Combined 3D Surface Plot of All Layers in a 3D NumPy Array

```python
import numpy as np
import plotly.graph_objects as go

# Example 3D array (3 layers, each 2x2)
three_dimension_vector = np.array([
    [[1, 2],
     [3, 4]],

    [[12, -9],
     [-3, 1]],

    [[-2, 7],
     [9, -4]]
])

num_layers, rows, cols = three_dimension_vector.shape

# Coordinates for surfaces
x = np.arange(cols)
y = np.arange(rows)
X, Y = np.meshgrid(x, y)

fig = go.Figure()

# Add each layer as a separate surface with a vertical offset
z_offset = 10  # distance between layers (increase/decrease if needed)

for i in range(num_layers):
    Z = three_dimension_vector[i] + i * z_offset  # shift each layer upward

    fig.add_trace(
        go.Surface(
            z=Z,
            x=X,
            y=Y,
            name=f"Layer {i}",
            showscale=(i == 0)  # show colorbar only once (minimal look)
        )
    )

fig.update_layout(
    title="Combined 3D Surface Plot (All Layers in One Figure)",
    scene=dict(
        xaxis_title="Column Index",
        yaxis_title="Row Index",
        zaxis_title="Values (with layer offset)"
    )
)

fig.show()
```
