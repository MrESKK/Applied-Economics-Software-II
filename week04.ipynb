#Optimization

In this section, the firm's profit maximization problem is introduced.  
Profit is defined as the difference between total revenue and total cost.

$$
\pi(Q) = R(Q) - C(Q)
$$

The cost function is given by:

$$
C(Q) = \frac{1}{3}Q^3 - 7Q^2 + 111Q + 50
$$

The demand function is:

$$
P = 100 - Q
$$

Therefore, the revenue function becomes:

$$
R(Q) = Q(100 - Q) = 100Q - Q^2
$$

Substituting these expressions into the profit function gives:

$$
\pi(Q) = -\frac{1}{3}Q^3 + 6Q^2 - 11Q - 50
$$

## Numerical Evaluation of the Profit Function

In this step, the profit function is evaluated for integer values of output \(Q\).  
This allows us to observe how profit changes as production increases.

```python
def profit(Q):
    return (-1/3)*Q**3 + 6*Q**2 - 11*Q - 50

for Q in range(0, 101):
    p = profit(Q)
    print("Q:", Q, "Profit:", p)
```

## Finer Search Using Decimal Intervals

In this step, the profit function is evaluated over smaller intervals of output.  
Instead of checking only integer values, the program searches over decimal values to obtain a more precise approximation of the profit-maximizing quantity.

```python
def profit(Q):
    return (-1/3)*Q**3 + 6*Q**2 - 11*Q - 50

for i in range(0, 10001):
    Q= i/10
    p = profit(Q)
    print("Q:", Q, "Profit:", p)
```

## Finding the Maximum Profit Numerically

This part of the notebook identifies the output level that generates the highest profit.  
The algorithm compares profit values over a grid of possible quantities and stores the best result.

$$
\pi(Q) = -\frac{1}{3}Q^3 + 6Q^2 - 11Q - 50
$$

```python
def profit(Q):
    return (-1/3)*Q**3 + 6*Q**2 - 11*Q - 50
best_q = 0
best_p = -10**18
for i in range(0, 1001):
    Q= i/10
    p = profit(Q)
    if p>best_p:
         best_p = p
         best_q = Q

print("Best Q:", best_q, "Best Profit:", best_p)
```

## Symbolic Solution Using Calculus

In addition to numerical search, the profit maximization problem can also be solved symbolically.  
Using calculus, the first-order condition is obtained by equating marginal revenue and marginal cost:

$$
MR = MC
$$

Where:

$$
TR = 100Q - Q^2
$$

$$
TC = \frac{1}{3}Q^3 - 7Q^2 + 111Q + 50
$$

The derivatives are:

$$
MR = \frac{dTR}{dQ}
\qquad
MC = \frac{dTC}{dQ}
$$

```python
import sympy as sp

# Sembol tanımla
Q = sp.symbols('Q')

# Fonksiyonlar
TC = (1/3)*Q**3 - 7*Q**2 + 111*Q + 50
TR = 100*Q - Q**2

# Türevler
MR = sp.diff(TR, Q)
MC = sp.diff(TC, Q)

# MR = MC denklemini çöz
q_star = sp.solve(sp.Eq(MR, MC), Q)

print("MR =", MR)
print("MC =", MC)
print("q* değerleri =", q_star)
```

# Two-Variable Profit Maximization

In this section, the profit maximization problem is extended to two decision variables.  
The objective is to determine the values of \(Q_1\) and \(Q_2\) that maximize total profit.

## Numerical Search for the Best Combination of \(Q_1\) and \(Q_2\)

The profit function is defined as:

$$
\pi(Q_1, Q_2) = 55Q_1 + 70Q_2 - 3Q_1Q_2 - 2Q_1^2 - 3Q_2^2
$$

A grid search method is used to test many possible combinations of \(Q_1\) and \(Q_2\), and the combination that yields the highest profit is selected.

```python
# Profit function
def profit(Q1, Q2):
    return 55*Q1 + 70*Q2 - 3*Q1*Q2 - 2*Q1**2 - 3*Q2**2

best_q1 = None
best_q2 = None
best_p = -10**18

for i in range(0,1001):
    Q1 = i/100
    for j in range(0,1001):
        Q2 = j/100

        p = profit(Q1, Q2)

        if p > best_p:
            best_p = p
            best_q1 = Q1
            best_q2 = Q2

print("Best Q1:", best_q1)
print("Best Q2:", best_q2)
print("Best Profit:", best_p)
```

# Three-Variable Profit Maximization

In this section, the optimization problem is extended to three decision variables.  
The aim is to find the values of \(Q_1\), \(Q_2\), and \(Q_3\) that maximize the firm's profit.

## First Three-Variable Example

The profit function is written as:

$$
\pi(Q_1, Q_2, Q_3) = 48Q_1 - 4Q_1^2 + 90Q_2 - 5Q_2^2 + 60Q_3 - 20
$$

The program searches over different values of the three variables and records the combination that produces the highest profit.

```python
def profit(Q1, Q2, Q3):
    return 48*Q1 - 4*Q1**2 + 90*Q2 - 5*Q2**2 + 60*Q3 - 20

best_q1 = None
best_q2 = None
best_q3 = None
best_p = -10**18

for i in range(0,101):
    Q1 = i/10
    for j in range(0,101):
      Q2 = j/10
      for k in range(0,101):
        Q3 = k/10

        p = profit(Q1, Q2, Q3)

        if p > best_p:
            best_p = p
            best_q1 = Q1
            best_q2 = Q2
            best_q3 = Q3
print("Best Q1:", best_q1)
print("Best Q2:", best_q2)
print("Best Q3:", best_q3)
print("Best Profit:", best_p)
```

## Second Three-Variable Example with Separate Revenue Components

In this example, each product has its own revenue function, and total cost depends on the sum of all outputs.

The revenue functions are:

$$
R_1 = 63Q_1 - 4Q_1^2
$$

$$
R_2 = 105Q_2 - 5Q_2^2
$$

$$
R_3 = 75Q_3 - 6Q_3^2
$$

The cost function is:

$$
C = 20 + 15(Q_1 + Q_2 + Q_3)
$$

Therefore, total profit is:

$$
\pi(Q_1, Q_2, Q_3) = R_1 + R_2 + R_3 - C
$$

The code then searches for the combination of output levels that maximizes total profit.

```python
def profit(Q1, Q2, Q3):
    r1 = 63*Q1 - 4*Q1**2
    r2 = 105*Q2 - 5*Q2**2
    r3 = 75*Q3 - 6*Q3**2
    c = 20 + 15*(Q1+Q2+Q3)
    return r1 + r2 + r3 - c

best_q1 = None
best_q2 = None
best_q3 = None
best_p = -10**18


for i in range(0,101):
    Q1 = i/10
    for j in range(0,101):
      Q2 = j/10
      for k in range(0,101):
        Q3 = k/10

        p = profit(Q1, Q2, Q3)

        if p > best_p:
            best_p = p
            best_q1 = Q1
            best_q2 = Q2
            best_q3 = Q3
print("Best Q1:", best_q1)
print("Best Q2:", best_q2)
print("Best Q3:", best_q3)
print("Best Profit:", best_p)
```
