A **CSV file** (Comma-Separated Values) is one of the most common file types used in data analysis.  
It stores data in table form, where each row is a record and each column is a variable.


# Reading Excel Files

Excel files are also widely used in economics, business, and social sciences.  
Pandas can read Excel files with the `read_excel()` function.



## Loading an Excel File

The file `world_population.xlsx` is read and stored in a DataFrame called `df_excel`.


```python

df_excel = pd.read_excel('/content/drive/MyDrive/Data/world_population.xlsx')
df_excel
```


## Selecting the Top 10 Countries by Rank

This example selects countries whose rank is less than or equal to 10, and then sorts them by rank.


```python

big_countries = df_excel[df_excel["Rank"] <= 10].sort_values("Rank")
big_countries
```


## Sorting the Same Countries Alphabetically

The same group of countries can also be sorted by country name instead of rank.


```python

alphabetical_countries = df_excel[df_excel["Rank"] <= 10].sort_values("Country")
alphabetical_countries
```


## Filtering Specific Countries by Name

The `isin()` function helps us check whether values belong to a list.  
Here, only Albania, Sweden, and Brazil are selected.


```python

specific_countries = df_excel[df_excel["Country"].isin(["Albania", "Sweden", "Brazil"])]
specific_countries
```


## Filtering Countries by Country Code

In many datasets, countries are represented by short codes.  
The following example selects countries according to the `CCA3` code.


```python

specific_countries_cca3 = df_excel[df_excel["CCA3"].isin(["USA", "GBR", "DEU", "TUR", "CZE"])]
specific_countries_cca3
```


## Searching for a Word Inside a Column

The `str.contains()` function is useful when we want to find all rows that contain a specific word.  
Below, the rows including the word **Republic** are displayed.


```python

df_excel[df_excel["Country"].str.contains("Republic", na=False)]
```


## Selecting Only Certain Columns

Sometimes we only need a few variables from a dataset.  
This example displays only the `Country` and `Capital` columns.


```python

df_excel[["Country", "Capital"]]
```


## Setting a Column as the Index

An index helps us identify rows more clearly.  
In this example, the `Country` column is used as the new index.


```python

df_excel_new_index = df_excel.set_index("Country")
df_excel_new_index
```


## Filtering Rows by Index Label

After setting `Country` as the index, we can search by row labels more easily.


```python

df_excel_new_index.filter(items=["Zimbabwe"], axis=0)
```


## Filtering Index Labels That Contain a Word

The `filter()` function can also be used with `like=` to find labels containing a certain expression.


```python

df_excel_new_index.filter(like="United", axis=0)
```


## Dropping an Unnecessary Column

Sometimes a dataset contains columns that are not needed for analysis.  
The following code removes the `Area (km¬≤)` column.


```python

df_excel_dropped = df_excel.drop(columns=["Area (km¬≤)"])
df_excel_dropped
```


## Checking the Shape of the DataFrame

The `shape` command tells us the number of rows and columns in a DataFrame.


```python

df_excel.shape
```


## Selecting Rows with a Condition

This example selects the row where the country rank is equal to 1.


```python

df_excel[df_excel["Rank"] == 1]
```


## Selecting Countries from a Specific Continent

Here, only European countries are shown.


```python

df_excel[df_excel["Continent"] == "Europe"]
```


## Using `.loc[]` for Row Selection

The `.loc[]` method is used to access rows by label.  
If the default index is numerical, it can also be used to display row ranges.


```python

df_excel.loc[10]
```

```python

df_excel.loc[0:10]
```


## Selecting Countries with Population Below 1 Million

The following code creates a smaller DataFrame that includes countries with a 2022 population below 1,000,000.


```python

df_population_less_than_1m = df_excel[df_excel["2022 Population"] < 1000000]
df_population_less_than_1m
```


## Selecting Certain Columns and Certain Rows Together

In this example, only the `Country` and `2022 Population` columns are selected, and only the first 11 rows are displayed.


```python

df_sub2022_pop = df_excel[["Country", "2022 Population"]].loc[0:10]
df_sub2022_pop
```
