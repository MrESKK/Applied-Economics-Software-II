#Main Matrix Definition

In this section, the main matrix \(A\) is defined using NumPy.  
The matrix represents the initial data structure that will be used for further operations such as slicing, submatrix extraction, and matrix multiplication.

$$
A =

4 & 11 & 20 & 1 & 9 \\
10 & 12 & 14 & 20 & 5 \\
8 & 1 & 20 & 18 & 30 \\
5 & 8 & 1 & 12 & 7
\end{bmatrix}
$$

```python
import numpy as np

A = np.array([[2,8,10,11,18],
              [4,11,20,1,9],
              [10,12,14,20,5],
              [8,1,20,18,30],
              [5,8,1,12,7]])

print(A)
```

The extracted submatrix represents a smaller portion of the original matrix and is stored for further analysis.

## Matrix Multiplication


In the final step, two matrices are multiplied using NumPy.

Matrix multiplication follows the mathematical rule:
$$
\textbf{We have two submatrices:}
$$

$$
B =
\begin{bmatrix}
12 & 14 & 20 \\
1 & 20 & 18
\end{bmatrix}
\qquad
C =
\begin{bmatrix}
2 & 8 & 10 \\
4 & 11 & 20 \\
10 & 12 & 14
\end{bmatrix}
$$

$$
\textbf{Dimensions}
$$

$$
B : 2 \times 3 \qquad
C : 3 \times 3
$$

$$
\text{Therefore, the matrix product } BC \text{ is defined and the result will be a } 2\times3 \text{ matrix.}
$$

---

$$
\textbf{Matrix Multiplication}
$$

First Row

$$
(12,14,20)
$$

$$
\begin{aligned}
a_{11} &= 12\cdot2 + 14\cdot4 + 20\cdot10 = 280 \\
a_{12} &= 12\cdot8 + 14\cdot11 + 20\cdot12 = 490 \\

Second Row

$$
(1,20,18)

## Solving a System of Linear Equations Using NumPy

Consider the following system of linear equations:

$$
2x + 8y + 10z = 18
$$

$$
4x + 11y + 20z = 9
$$

$$
10x + 12y + 14z = 5
$$

This system can be written in matrix form as:

$$
B \cdot V = e
$$

Where:

$$
B =
\begin{bmatrix}
2 & 8 & 10 \\
4 & 11 & 20 \\
10 & 12 & 14
\end{bmatrix}
$$

$$
V =
\begin{bmatrix}
x \\
y \\
z
\end{bmatrix}
$$

$$
e =
\begin{bmatrix}
18 \\
9 \\
5
\end{bmatrix}
$$

---

### Why do we separate the system as B, V, and e?

- **B** is the coefficient matrix.  
  It contains the numerical coefficients of the variables.

- **V** is the variable (unknown) vector.  
  It represents the variables we want to solve for (x, y, and z).

- **e** is the constant vector.  
  It contains the values on the right-hand side of the equations.

Writing the system in matrix form allows us to apply linear algebra methods instead of solving the equations manually.

---

### Mathematical Solution Concept

If matrix \(A\) is invertible, the solution of the system can be written as

$$
V = C^{-1} e
$$

Instead of explicitly computing the inverse matrix, numerical libraries such as NumPy provide more stable and efficient algorithms to compute the solution.

For example, NumPy uses functions such as

$$
\texttt{np.linalg.solve(C, E)}
$$
or
$$
\texttt{np.dot(np.linalg.inv(C), E)}
$$

These methods internally rely on efficient matrix factorization techniques ([such as LU decomposition](https://en.wikipedia.org/wiki/LU_decomposition)) to compute the solution.

---
This matrix-based approach is widely used in scientific computing, econometrics, optimization, and machine learning because it allows large systems of equations to be solved efficiently.
---
### Video Example

For an additional example of solving a system of linear equations using NumPy, you can [watch](https://www.youtube.com/watch?v=6eXl7sNZC9c) the following tutorial.



```python
Ve = np.linalg.solve(C,E)
Ve
```

```python
V = np.linalg.inv(C) @ E
print(V)
```

# Python Functions

In Python, a **function** is a reusable block of code designed to perform a specific task.

A function usually takes **parameters** as input, processes them inside the function body, and may produce a **return value** as output.

This can be represented conceptually as:

$$
\text{parameters} \;\longrightarrow\; \text{function} \;\longrightarrow\; \text{return value}
$$

In the visual representation:

- **Function** refers to the defined block such as `def function():`
- **Parameters** are the input values given to the function
- **Code** represents the instructions written inside the function body
- **Return value** is the output produced after the code is executed

Functions are important in Python because they help make programs more **organized**, **readable**, and **reusable**.

### General Example

Here:

- \(x\) is the input

# If–Else-Elif Statements in Python

This section demonstrates how **if–else-elif statements** work in Python using simple code examples.

Conditional statements allow a program to make decisions based on logical conditions.  
Depending on whether a condition evaluates to **True** or **False**, the program executes different blocks of code.

These examples illustrate how Python uses `if`, `elif`, and `else` to control the flow of execution and implement basic decision-making logic.

Understanding conditional statements is fundamental in programming because they allow programs to respond dynamically to different inputs and situations.
### Supplementary Video Resource

For a visual explanation of **conditional statements (`if`, `elif`, `else`) in Python**, the following tutorial provides an [additional demonstration](https://www.youtube.com/watch?v=hf-amsrHa4M).



```python
#This example demonstrates how Python uses conditional statements to evaluate a variable and execute different code blocks depending on the result.
temperature = 16
if temperature >= 30:
    print("It's hot outside!")
elif temperature <30 and temperature > 15:               # or elif 15 < temperature < 30
    print("It's a moderate day.")                        # The entire expression evaluates to True only if both conditions are true
elif temperature <= 15:                                  # "and" means In Python, the `and` operator is used to combine multiple conditions in a single statement.
    print("It's a cool day.")

```

### Interactive Python Program Example

This program is a simple interactive Python example.  
It asks the user for their **name** and then asks them to choose an **energy drink**.  
The program uses `input()` to receive information from the user and `print()` to display messages.

Using **conditional statements (`if`, `elif`, `else`)**, the program checks the user's choice and displays the corresponding response.

### Program Flow Diagram

$$
\begin{aligned}
\text{Start} \\
\downarrow \\
\text{User enters name} \\
\downarrow \\
\text{User selects a drink} \\
\downarrow \\
\text{Program checks the choice} \\
\downarrow \\
\text{Correct message and price displayed} \\
\downarrow \\
\text{End}
\end{aligned}
$$

```python
print("Welcome to the Space Station Supply Terminal!")
name = input("Astronaut, what is your name? ")

if name == "Unknown":
  print("Access denied. Identity verification failed.")
  exit()
else:
  print("Greetings", name, ".")
  print("Please choose your energy supply.")
  print("Available options: Plasma Drink, Nano Water, Cosmic Juice, Quantum Tea")
  drink = input()

if drink == "Plasma Drink":
  print("Energy levels boosted. Cost: 5 space credits.")
elif drink == "Nano Water":
  print("Hydration complete. Cost: 3 space credits.")
elif drink == "Cosmic Juice":
  print("Flavor detected: Stellar Mango. Cost: 6 space credits.")
elif drink == "Quantum Tea":
  print("Temporal relaxation activated. Cost: 4 space credits.")
else:
  print("Error:", drink, "is not available in the station inventory.")
```
